```python
import pandas as pd

# Load the dataset
data = pd.read_csv('../data/pokemon_gen_1-2.csv')

# Assuming we class the pokemon as light, medium, heavy based on the weight first 33% is light, 33-66% is medium, 66-100% is heavy
itWeighsdict = {
    'light': data.weight.quantile(0.33),
    'heavy': data.weight.quantile(0.66)
}

itWeighsList = ["itWeighs light", "itWeighs medium", "itWeighs heavy"]

# Assuming we class the pokemon as short, medium, tall based on the height first 33% is short, 33-66% is medium, 66-100% is tall
itHasHeight = {
    'short': data.height.quantile(0.33),
    'tall': data.height.quantile(0.66)
}

itHasHeightList = ["itHasHeight short", "itHasHeight medium", "itHasHeight tall"]

list_of_types1 = data.type_1.unique()
type1list = ["isType " + l for l in list_of_types1]
print("-------Type 1 List-------")
print(type1list)
print("\n\n")
print("-------Type 2 List-------")
list_of_types2 = [x for x in data.type_2.unique() if str(x) != 'nan']
type2list = ["isType " + l for l in list_of_types2]
print(type2list)
print("\n\n")
print("-------Genus List-------")
genus = [x.split()[:-1] for x in data.genus.unique()]
genus = [''.join(x) for x in genus]
genusList = ["isGenus " + g for g in genus]
print(genusList)
print("\n\n")
print("-------Gen Introduced In List-------")
genIntroduceInList = ["wasIntroducedIn Gen" + str(g) for g in data.gen_introduced.unique()]
print(genIntroduceInList)
print("\n\n")
print("-------Legendary or Not-------")
legendaryornot = ['isLegendary']
print(legendaryornot)
print("\n\n")
print("-------Can Evolve-------")
canEvolve = ['canEvolve']
print(canEvolve)
print("\n\n")
print("-------Evolves From-------")
evolvesFrom = [x for x in data.evolves_from.unique() if str(x) != 'nan']
evolvesFromList = ["evolvesFrom " + e for e in evolvesFrom]
print(evolvesFromList)
print("\n\n")
print("-------Primary Color-------")
primaryColorList = ["isColor " + p for p in data.primary_color.unique()]
print(primaryColorList)
print("\n\n")
print("-------Shape-------")
shapeList = ["hasShape " + s for s in data['shape'].unique()]
print(shapeList)
print("\n\n")
for col in data.columns:
    if 'attack_effectiveness' in col:
        howEffective = data[col].unique()
        howEffectiveList = []
        for h in howEffective:
            # if h is NaN
            if str(h)== 'nan':
                howEffectiveList.append("howEffective"+col.split("_")[0].capitalize()+" "+"normal")
            if "super" in str(h):
                howEffectiveList.append("howEffective"+col.split("_")[0].capitalize()+" "+"superEffective")
            elif "not_very" in str(h):
                howEffectiveList.append("howEffective"+col.split("_")[0].capitalize()+" "+"notVeryEffective")
        howEffectiveList= howEffectiveList[:3]
        print("--- how effective is "+col.split("_")[0].capitalize()+" attack against this pokemon ---")
        print(howEffectiveList)
        print("\n\n")
```
