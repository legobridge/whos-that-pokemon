```python
import pandas as pd

# Load the dataset
data = pd.read_csv('../data/PokemonDataLatest.csv')

print("------It Weighs-------")
itWeighsdict = {
    'veryLight': 50,
    'veryHeavy': 1000,
}

itWeighsList = ["itWeighs veryLight", "itWeighs veryHeavy"]
print(itWeighsList)
print("\n\n")
print("------It Has Height-------")
itHasHeight = {
    'veryShort': 7,
    'veryTall': 25
}

itHasHeightList = ["itHasHeight veryShort", "itHasHeight veryTall"]
print(itHasHeightList)
# print("\n\n")
list_of_types1 = data.type_1.unique()
type1list = ["isType " + l for l in list_of_types1]
# print("-------Type 1 List-------")
# print(type1list)
# print("\n\n")
# print("-------Type 2 List-------")
list_of_types2 = [x for x in data.type_2.unique() if str(x) != 'nan']
type2list = ["isType " + l for l in list_of_types2]
# print(type2list)
print("\n\n")
print("-------Type List-------")
typeList = list(set(type1list+type2list))
print(typeList)
print("\n\n")
print("-------Genus List-------")
genus = [x.split()[:-1] for x in data.genus.unique()]
genus = [''.join(x) for x in genus]
genusList = ["isGenus " + g for g in genus]
print(genusList)
print("\n\n")
print("-------Gen Introduced In List-------")
genIntroduceInList = ["wasIntroducedIn Gen" + str(g) for g in data.gen_introduced.unique()]
print(genIntroduceInList)
print("\n\n")
print("-------Legendary or Not-------")
legendaryornot = ['isLegendary']
print(legendaryornot)
print("\n\n")
print("-------Can Evolve-------")
canEvolve = ['canEvolve']
print(canEvolve)
print("\n\n")
print("-------first letter of pokemon list-------")
firstLetterList = ["isFirstLetter " +  chr(i) for i in range(65,91)]
print(firstLetterList)
print("\n\n")
print("-------Evolves From-------")
evolvesFrom = [x for x in data.evolves_from.unique() if str(x) != 'nan']
evolvesFromList = ["evolvesFrom " + e for e in evolvesFrom]
print(evolvesFromList)
print("\n\n")
print("-------Evolves From Something-------")
evolvesFromSomething = ['evolvesFromSomething']
print(evolvesFromSomething)
print("\n\n")
print("-------Color-------")
colorList = ["isColor " + p for p in data.primary_color.unique()]
print(colorList)
print("\n\n")
print("-------Shape-------")
shapeList = ["hasShape " + s for s in data['shape'].unique()]
print(shapeList)
print("\n\n")
print("-------Ash Owned Pokemon-------")
isAshOwnedList = ['isAshOwned']
print(isAshOwnedList)
print("\n\n")
print("-------IsPrimaryColor-------")
primaryColors = ["red","blue","yellow"]
isPrimaryColorList = ["isPrimaryColor"]
print(isPrimaryColorList)
print("\n\n")
print("-------Attack Effectiveness-------")
effectiveList = list(list())
effectivenessTypeList=['super_effective','not_very_effective','has_no_effect']
for col in data.columns:
    if 'attack_effectiveness' in col:
        howEffectiveList=[]
        for h in effectivenessTypeList:
            howEffectiveList.append("howEffective " + col.split("_")[0].capitalize() + " " + h)
        effectiveList.append(howEffectiveList)
        print("--- how effective is "+col.split("_")[0].capitalize()+" attack against this pokemon ---")
        print(howEffectiveList)
        print("\n\n")
print("-------Pokemon name-------")
pokemonNameList = ["Pokemon"]
print(pokemonNameList)
print("\n\n")
```

```python
## Create a sparse matrix for setting up Pokémon rules

new_df=pd.DataFrame(columns=pokemonNameList+itWeighsList+itHasHeightList+typeList+genusList+genIntroduceInList+legendaryornot+canEvolve+evolvesFromList+firstLetterList+evolvesFromSomething+colorList+shapeList+isAshOwnedList+isPrimaryColorList+[item for sublist in effectiveList for item in sublist])
new_df['Pokemon']=data['name']
for ele in typeList:
    new_df[ele]=data.apply(lambda x: 1 if ele.split(" ")[1] in [x['type_1'],x['type_2']] else 0,axis=1)
for ele in itWeighsList:
    if ele=="itWeighs veryLight":
        new_df[ele]=data['weight'].apply(lambda x: 1 if int(x)<int(itWeighsdict["veryLight"]) else 0)
    elif ele=="itWeighs veryHeavy":
        new_df[ele]=data['weight'].apply(lambda x: 1 if int(x)>int(itWeighsdict["veryHeavy"]) else 0)
for ele in itHasHeightList:
    if ele=="itHasHeight veryShort":
        new_df[ele]=data['height'].apply(lambda x: 1 if int(x)<int(itHasHeight["veryShort"]) else 0)
    elif ele=="itHasHeight veryTall":
        new_df[ele]=data['height'].apply(lambda x: 1 if int(x)>int(itHasHeight["veryTall"]) else 0)
for ele in genusList:
    new_df[ele]=data['genus'].apply(lambda x: 1 if ''.join(x.split(" ")[:-1])==ele.split(" ")[1] else 0)
for ele in firstLetterList:
    new_df[ele]=data['name'].apply(lambda x: 1 if x[0].lower()==ele.split(" ")[1].lower() else 0)
for ele in genIntroduceInList:
    new_df[ele]=data['gen_introduced'].apply(lambda x: 1 if str(x)==str(ele.split(" ")[1][-1:]) else 0)
for ele in legendaryornot:
    new_df[ele]=data['legendary'].apply(lambda x: 1 if x==True else 0)
for ele in canEvolve:
    new_df[ele]=data['can_evolve'].apply(lambda x: 1 if x==True else 0)
for ele in evolvesFromList:
    new_df[ele]=data['evolves_from'].apply(lambda x: 1 if x==ele.split(" ")[1] else 0)
for ele in evolvesFromSomething:
    new_df[ele]=data['evolves_from'].apply(lambda x: 1 if str(x)!='nan' else 0)
for ele in colorList:
    new_df[ele]=data['primary_color'].apply(lambda x: 1 if x==ele.split(" ")[1] else 0)
for ele in shapeList:
    new_df[ele]=data['shape'].apply(lambda x: 1 if x==ele.split(" ")[1] else 0)
for ele in shapeList:
    new_df[ele]=data['shape'].apply(lambda x: 1 if x==ele.split(" ")[1] else 0)
for ele in shapeList:
    new_df[ele]=data['shape'].apply(lambda x: 1 if x==ele.split(" ")[1] else 0)
for ele in isPrimaryColorList:
    new_df[ele]=data['primary_color'].apply(lambda x: 1 if x.lower() in primaryColors else 0)
#flatten 2d howEffectiveList
howEffectiveList = [item for sublist in effectiveList for item in sublist]
for ele in howEffectiveList:
    new_df[ele]=data[ele.split(' ')[1].lower()+"_attack_effectiveness"].apply(lambda x: 1 if ele.split(" ")[2] in str(x) else 0)
ash_df = pd.read_csv('../data/PokemonOwnedbyAsh.csv')
for ele in isAshOwnedList:
    new_df[ele]=data['name'].apply(lambda x: 1 if x in ash_df['Pokemon'].values else 0)
```

```python
# create rules from the sparse matrix for each Pokémon
# for example, rule: ((isColor Yellow) (isLegendary) … (isType Electric)) -> (isPokemon Zapdos)
rules = []
for i in range(len(new_df)):
    s = "("
    for j in range(len(new_df.columns)):
        if new_df.iloc[i,j] == 1 and new_df.columns[j] not in s:
            s += "("+new_df.columns[j]+")" + " "
    s+=")" + " -> " + "(" + "isPokemon " + new_df.iloc[i,0].replace(" ", "") + ")"
    rules.append(s)
rules
```

```python
# write rules to a file
with open('../data/rules_generated.txt', 'w') as f:
    for item in rules:
        f.write("rule: %s\n" % item)
```
